```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

# One to many relationship

# 1. Report the account representative for each customer.

```python
run("""
SELECT
    customerNumber,
    customerName,
    CONCAT(firstName, ' ', lastName) AS representativeName,
    salesRepEmployeeNumber as representativeNumber
FROM
    customers c
        INNer JOIN
    employees e ON c.salesRepEmployeeNumber = e.employeeNumber
""")
```

# 2. Report total payments for Atelier graphique.

```python
run("""
SELECT 
    customerNumber, 
    customerName, 
    SUM(amount) AS totalPayments
FROM
    customers
        INNER JOIN
    payments USING(customerNumber)
WHERE customerNumber = 103
""")
```

# 3. Report the total payments by date

```python
run("""
SELECT 
    paymentDate,
    SUM(amount) AS totalPayments
FROM
    customers
        INNER JOIN
    payments USING(customerNumber)
GROUP BY paymentDate
""")
```

# 4. Report the products that have not been sold.

```python
run("""
SELECT 
    productName,
    orderNumber AS statusSelling
FROM
    products t1
        LEFT JOIN
    orderdetails t2 ON t1.productCode = t2.productCode
WHERE
    orderNumber IS NULL
""")
```

# 5. List the amount paid by each customer.

```python
run("""
SELECT 
    customerNumber,
    CustomerName,
    SUM(amount) AS totalPayments
from 
    customers
        LEFT JOIN
    payments USING(customerNumber)
GROUP BY 
    customerNumber
""")
```

# 6. How many orders have been placed by Herkku Gifts?

```python
run("""
SELECT
    customerName,
    COUNT(orderNumber) AS totalOrder
FROM
    orders o
        INNER JOIN
    customers c ON o.customerNumber=c.customerNumber
WHERE customerName="Herkku Gifts"
""")
```

# 7. Who are the employees in Boston?

```python
run("""
SELECT
    CONCAT(firstName, ' ' ,lastName) as employeeName,
    city
FROM
    employees 
        INNER JOIN
    offices USING (officeCode)
WHERE city='boston'
""")
```

# 8. Report those payments greater than $100,000. Sort the report so the customer who made the highest payment appears first.

```python
run("""
SELECT
    customerNumber,
    customerName,
    SUM(amount) AS totalAmount
FROM
    customers
        INNER JOIN
    payments USING (customerNumber)
GROUP BY customerNumber
HAVING totalAmount > 100000
ORDER BY totalAmount DESC
""")
```

# 9. List the value of 'On Hold' orders.

```python
run("""
SELECT 
    productCode,
    priceEach,
    quantityOrdered as 'Quantity', 
    status
FROM 
    orderdetails
        INNER JOIN
    orders USING(orderNumber)
WHERE 
    status='On Hold'
""")
```

# 10. Report the number of orders 'On Hold' for each customer.

```python
run("""
SELECT 
    customerName,
    COUNT(status) AS 'Number of orders hold'
FROM
    customers c
        INNER JOIN
    orders o 
        ON c.customerNumber = o.customerNumber
WHERE 
    status='On Hold'
GROUP BY 
    customerName
""")
```

# Many to many relationship

# 1. List products sold by order date.

```python
run("""
SELECT 
    productName,    
    orderDate
FROM
    ((orderdetails t1
        INNER JOIN
    orders t2 ON t1.orderNumber = t2.orderNumber)
        INNER JOIN
    products t3 ON t1.productCode = t3.productCode)  
ORDER BY
    orderDate 
LIMIT 0,20
""")
```

# 2. List the order dates in descending order for orders for the 1940 Ford Pickup Truck.

```python
run("""
SELECT 
    productName,    
    orderDate
FROM
    ((orderdetails t1
        INNER JOIN
    orders t2 ON t1.orderNumber = t2.orderNumber)
        INNER JOIN
    products t3 ON t1.productCode = t3.productCode)  
WHERE productName='1940 Ford Pickup Truck'
ORDER BY
    orderDate DESC
LIMIT 0,20
""")
```

# 3. List the names of customers and their corresponding order number where a particular order from that customer has a value greater than /$25,000?

```python
run("""
SELECT 
    customerName,
    orderNumber,
    amount
FROM
    ((customers t1
        INNER JOIN
    orders t2 ON t1.customerNumber = t2.customerNumber)
        INNER JOIN
    payments t3 ON t1.customerNumber = t3.customerNumber)  
 WHERE amount>25000
LIMIT 0,20
""")
```

# 4. Are there any products that appear on all orders?

```python
run("""
SELECT 
    productName,
    COUNT(productCode) AS numberOfAppearanceOnOrders
FROM 
    orderdetails 
        INNER JOIN
    products USING(productCode)
GROUP BY 
    productCode 
HAVING 
    numberOfAppearanceOnOrders = (SELECT COUNT(*) FROM orders)
""")
```

# 5. List the names of products sold at less than 80% of the MSRP.

```python
run("""

SELECT  
    p.productName,
    (d.priceEach*100/p.MSRP) AS less80percentMSRP
FROM 
    products p
    INNER JOIN 
    orderdetails d USING(productCode)
WHERE 
    (d.priceEach*100/p.MSRP) < 80
""")
```

# 7. List the products ordered on a Monday.

```python
run("""
SELECT 
    productCode,
    productName,
    orderNumber,
    orderDate,
    DAYNAME(orderDate)
FROM 
    products
        INNER JOIN 
    orderdetails od USING (productCode)
        INNER JOIN 
    orders o USING (orderNumber)
WHERE 
    DAYNAME(orderDate) = 'Monday'
LIMIT 0,10
""")
```

# 8. What is the quantity on hand for products listed on 'On Hold' orders?

```python
run("""
SELECT 
    productCode,
    productName,
    SUM(quantityOrdered) as quantityOnHand,
    status
FROM 
    orders o
        INNER JOIN 
    orderdetails od USING (orderNumber)
        INNER JOIN 
    products p USING (productCode)
WHERE status = 'On Hold'
GROUP BY productCode
""")
```
