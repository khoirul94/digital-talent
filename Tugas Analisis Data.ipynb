# Tugas Analisis DAta - Data Angkutan Umum DKI Jakarta

# Deskripsi Data

Data yang diambil berisi mengenai data angkutan umum di DKI Jakarta

Penjelasan mengenai Variabel pada Dataset ini:

1. operator: Operator kendaraan umum
2. jenis_kendaraan: Jenis kendaraan
3. jenis_layanan: Jenis Layanan
4. jumlah_armada: Jumlah dari armada
5. jumlah_trayek: Jumlah trayek

Data bersumber dari Pemerintah Provinsi DKI Jakarta, 2013.

Sumber data : https://data.go.id/dataset/data-angkutan-umum-dki-jakarta-2

```python
import pandas as pd
data_angkutan_umum = pd.read_csv('processed-data-kendaraan-angkutan-umum-dki-jakarta-tahun-2013.csv', delimiter=",", encoding="utf-8")
```

```python
data_angkutan_umum
```

```python
# Load data daftar operator angkutan umum di jakarta berdasarkan jenis kendaraan Bus Kota

data_angkutan_umum[data_angkutan_umum['jenis_kendaraan']=='Bus Kota']
```

# Statistik Data

```python
#jumlah armada dan jumlah trayek angkutan umum di Jakarta

armada = data_angkutan_umum.jumlah_armada.sum()
trayek = data_angkutan_umum.jumlah_trayek.sum()

print("Jumlah armada angkutan umum di Jakarta : ", armada)
print("Jumlah Trayek angkutan umum di Jakarta : ", trayek)
```

```python
# Jumlah angkutan umum di Jakarta berdasarkan jenis kendaraan

data_angkutan_umum.groupby('jenis_kendaraan').sum()
```

```python
# Jumlah angkutan umum di Jakarta berdasarkan jenis layanan

data_angkutan_umum.groupby('jenis_layanan').sum()
```

```python
#Total armada tiap operator angkutan umum

data_angkutan_umum.groupby(['operator', 'jenis_kendaraan', 'jenis_layanan']).sum()['jumlah_armada']
```

```python
# Operator dengan jumlah armada paling banyak

armada_max = data_angkutan_umum.jumlah_armada.max()
data_angkutan_umum[data_angkutan_umum['jumlah_armada'] == armada_max]
```

```python
# Operator dengan jumlah armada paling sedikit

armada_min = data_angkutan_umum.jumlah_armada.min()
data_angkutan_umum[data_angkutan_umum['jumlah_armada'] == armada_min]
```

```python
# Mean dari jumlah armada

mean = data_angkutan_umum.jumlah_armada.mean()
print('Hasil (mean) :',mean)
```

```python
# Mean dari jumlah armada

median = data_angkutan_umum.jumlah_armada.median()
print('Hasil (median) :',median)
```

# Visualisasi dan Analis Data

```python
# Jumlah armada tiap operator

ax = data_angkutan_umum.groupby('operator').sum()['jumlah_armada']
ax = ax.plot(kind='bar', title ="Jumlah Armada Angkutan Umum Per Operator (2013)", figsize=(20,6), legend=True, fontsize=12)
ax.set_xlabel("Nama Operator Angkutan Umum", fontsize=12)
ax.set_ylabel("Jumlah Armada", fontsize=12)
```

Berdasarkan total data angkutan umum di Jakarat tahun 2013 dapat dilihat bahwa operator angkutan umum yang memiliki armada paling banyak yaitu Mayasari Bakti, disusul Perum PPD di urutan ke dua dan PT Koda Jaya/AJA Putra di urutan ke tiga.

```python
# Rata-rata Angkutan Umum berdasarkan jenis kendaran

ax = data_angkutan_umum.groupby('jenis_kendaraan').mean()['jumlah_armada']
ax = ax.plot(kind='bar', title ="Rata-Rata Angkutan Umum berdasarkan Jenis kendaraan", legend=True, fontsize=12)
ax.set_xlabel("Jenis Kendaraan", fontsize=12)
ax.set_ylabel("Armada", fontsize=12)
```

```python
#Rata-Rata Angkutan Umum berdasarkan Jenis kendaraan

data_angkutan_umum.groupby('jenis_kendaraan').mean()['jumlah_armada']
```

Berdasarkan data angkutan umum di Jakarat tahun 2013 dapat dilihat bahwa rata-rata jenis kendaraan didominasi paling banyak bus kota.

```python
# Rata-rata Angkutan Umum berdasarkan jenis layanan

ax = data_angkutan_umum.groupby('jenis_layanan').mean()['jumlah_armada']
ax = ax.plot(kind='bar', title ="Rata-Rata Angkutan Umum berdasarkan Jenis layanan", legend=True, fontsize=12)
ax.set_xlabel("Jenis layanan", fontsize=12)
ax.set_ylabel("Armada", fontsize=12)
```

```python
#Rata-Rata Angkutan Umum berdasarkan Jenis layanan

data_angkutan_umum.groupby('jenis_layanan').mean()['jumlah_armada']
```

Berdasarkan data angkutan umum di Jakarat tahun 2013 dapat dilihat bahwa rata-rata jenis layanan didominasi paling banyak oleh bus yang memiliki layanan Patas AC.
